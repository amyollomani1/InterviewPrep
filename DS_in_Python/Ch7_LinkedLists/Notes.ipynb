### Disadvantages to arrays

1. Length of dynamic array might be longer than acutal number of elements it stores
2. Amortized bounds for operations may be unacceptable in real-time solutions
3. Insertions and deletions at interior positions of array are expensive

Arrays use one large chunk of memory
A linked list uses nodes allocated for each element. Each node maintains a reference to its element, and 1+ references to neighboring nodes to collectivly represent linear order of sequence

### Single Linked List
Its a collection of nodes forming a linaer sequence. Each node stores reference to an object and a reference to next node in list

First and last node are head and tail
Tail has None as its next reference
Pointer hopping = traversing linked list

Linked list doesn't have predertemined fixed size. It uses space proprotional to current num of elements

    -newest = Node(e)

```python
# top of stack is head of list

class LinkedStack:
    #-------------------------- nested Node class --------------------------
    class _Node:
        __slots__ = '_element', '_next' #streamline memory usage
        
        def __init__(self, element, next):
            #two instance variables
            self._element = element
            self._next = next
        
    #------------------------------- stack methods -------------------------------
    def __init__(self):
    #”””Create an empty stack.””” 
        self._head = None
        self._size = 0
    # reference to the head node # number of stack elements
    def __len__(self):
    #”””Return the number of elements in the stack.””” 
        return self._size
    def is_empty(self):
    #”””Return True if the stack is empty.””” 
        return self._size == 0
    def push(self, e):
    #”””Add element e to the top of the stack.”””
        self._head = self._Node(e, self._head) # create and link a new node self. 
        size += 1
    
    def top(self):
#”””Return (but do not remove) the element at the top of the stack.
#Raise Empty exception if the stack is empty. ”””
        if self.is_empty():
            raise Empty( 'Stack is empty' )
        return self._head._element # top of stack is at head of list
    
    def pop(self):
        #Remove and return elm at top of stack
        
        if self.is_empty():
            raise Empty('Stack is Empty')
        
        answer = self._head._element
        self._head = self._head._next #self._head is reference to first node of linked list
        size -=1
        return answer
        
        
```

### Implementing Queue with Singly Linked List

Worst case for all operations are O(1)

head and tail reference as instance variables
front of queue is head, back is tail of list.
FIFO

This code is pretty similar to LinkedStack excpet for use of tail node
    all operations worst-case is O(1) and space is linear to current number of elements


```python
class LinkedQueue:
    #-------------------------- nested Node class --------------------------
    class _Node:
        def __init__(self, element, next):
            #two instance variables reference to object and reference to next node
            self._element = element
            self._next = next
        
    #------------------------------- stack methods -------------------------------
    def __init__(self):
    #”””Create an empty stack.””” 
        self._head = None
        self._tail = None
        self._size = 0
    # reference to the head node # number of stack elements
    def __len__(self):
    #”””Return the number of elements in the stack.””” 
        return self._size
    def is_empty(self):
    #”””Return True if the stack is empty.””” 
        return self._size == 0
    
    def first(self):
        #return Don't remove elm at front of queue
        if self.is_empty():
            raise Empty('Queue is empty')
        return self._head._element
    
    def dequeue(self):
        #remove and return first elm (FIFO)
        if self.is_empty():
            raise Empty('Queue is empty')
        answer = self._head._element
        self._head = self._head._next
        self._size -=1
        if self.is_empty():
            self._tail = None #specail case if queue is empty: removed head had been the tail
        return answer
        
    def push(self, e):
    #”””Add element e to the top of the stack.”””
        self._head = self._Node(e, self._head) # create and link a new node self. 
        size += 1
        
    def enequeue(self,e):
        #add elm to back of queue
        newest = self._Node(e,None) #node will be new tail node
        if self.isempty():
            self._head = newest
        else:
            self._tail._next = newest
        self._tail = newest #update reference to node
        self._size += 1
        
```

### Circularly Linked Lists

Tail of the list uses its next reference to point to head of list
* Circular arrays were just an illusion, they weren't actually circular
* since theres no beginning an array, need to use current reference to describe a designated node. 

Traverse nodes using current = current.next

Round-Robin Scheduler allocates CPU time slices
1. e = Q.dequeue()
2. Service elmenent e
3. Q.enequeu(e) put it at the end again

* this is alot of work. Using a circularQueue, you can just use rotate() which moves first elm of queue to back
1. Service element Q.front()
2. Q.rotate()

Instance Variables:
* _tail: refrence to tail node (None when empty)
* _size: curr number of elements in queue
 
Note: self._tail._next = head of queue
enqueue(): places new node just after tail but before current head, then new node becomes the tail
rotate(): removes front elment and reinserts it at back of queue


```python
class CircularQueue:
    
    class _Node:
        def __init__(self, element, next):
            #two instance variables reference to object and reference to next node
            self._element = element
            self._next = next
    
    def __init__(self):
        self._tail = None
        self._size = 0
    
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    def first(self):
        #return but don't remove first elm of queue 
        if self.is_empty():
            return Empty('Queue is empty')
        head = self._tail._next   
        return head._element
    
    def dequeue(self):
        #removes and returns first elm of queue (FIFO)
        if self.is_empty():
            return Empty('Queue is empty')
        oldhead = self._tail._next
        if self._size == 1:
            self._tail = None #queue becomes empty since there was only 1 elm
        else:
            self._tail._next = self._tail._next._next #or =  oldhead._next
        self._size -=1
        return oldhead._element
    
    def enqueue(self,e):
        #add elm to back of queue
        newest = self._Node(e,None) #node will be new tail node
        if self.is_empty():
            newest._next = newest
        else:
            newest._next = self._tail._next #new node points to head
            self._tail._next = newest #old tail points to new node
            
        self._tail = newest #tail becomes new
        self._size += 1
        
    def rotate(self):
        #rotate front elm to back of queue
        if self._size > 0:
            self._tail = self._tail._next #old head becomes new tail
```

### Doubly Linked Lists

Cons to single linked list: can efficinelt insert node to beginning or end, and delet a node at head, but can't efficient delete a node at tail, or efficiently delete a node in the middle because you can't determine node that prcedes the node

Contains a "header" and "trailer" node at beginning and end (sentinels). These nodes containe no elements, so realy first elm has a first node to refer to etc.

Pros to sentinels:
- Extra space devoted greatly simplifies logic of operations, and header and trailer nodes never change.

#### Time Complexity
- Insertions and deletions: O(1) time worst case, only if location of an opeation can be succinctly identified
    - To find jth elm and preform operations on it, you would have to traverse that portion of the list


```python
class _DoublyLinkedBase:
    class _Node:
        __slots__ = '_element', '_prev', '_next' #streamlines memory
        
        def __init__(self, element, prev, next):
            self._element = element
            self._prev = prev #previous node refrence
            self._next = next #next node reference
            
    def __init__(self):
        self._header = self._Node(None, None, None)
        self._trailer = self._Node(None, None, None)
        self._header._next = self._trailer
        self._trailer._prev = self._header
        self._size = 0 #number of elements
        
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def _insert_between(self, e, predecessor, successor):
        # add elm e between 2 exisitng nodes and return new node
        newest = self._Node(e, predecessor, successor) #linked to neighbors
        predecessor._next = newest
        successor._prev = newest
        self._size +=1
        return newest
    
    def _delet_node(self,node):
        #delet nonsentinel node from list and return elm
        predecessor = node._prev
        sucessor = node._next
        predecessor._next = sucessor
        sucessor._prev = predecessor
        self._size -=1
        element = node._element
        node._prev = node._next = node._element = None #deprecate node
        return element
    
class Linked(_DoublyLinkedBase):
    
    def first(self):
        #return but don't remove first elm at front of deque
        if self.is_empty():
            raise ("Deque is empty")
        return self._header._next._element
    
    def last(self):
        if self.is_empty():
            raise ("Deque is empty")
        return self._tailer._prev._element
    
    def insert_first(self,e):
        self._insert_between(e, self._header, self._header._next)
        
    def insert_last(self, e):
        self._insert_between(e, self._tailer._prev, self._tailer)
    
    def delete_first(self):
        #remove and return elm from front of deque
        if self.is_empty():
            raise("Deque empty")
        return self._delete_node(self._header._next)
    
    def delete_last(self):
        if self.is_empty():
            raise("Deque empty")
        return self._delete_node(self._tailer._prev)
    
                
```

Why we can't use indecies when working with linked lists:

```python
class PositionalList(_DoublyLinkedBase):
    
    class Position:
        def __init__(self, container, node):
            self._container = container
            self._node = node
            
        def element(self):
            return self._node._element
        
        def __eq_(self,other):
            #return true if other is a position representing same location
            return type(other) is type(self) and other._node is self._node
        
        def __ne__(self, other):
            #return true if other doesn't represent same location
            return not(self == other)
    
    def _validate(self, p):
        #returns position's node, or raise error
        if not isinstance(p, self.Position):
            raise TypeError( p must be proper Position type ) 
        if p. container is not self:
            raise ValueError( p does not belong to this container )
        if p. node. next is None: # convention for deprecated nodes
            raise ValueError( p is no longer valid )
        return p._node
    
    def _make_position(self,node):
        #return position instance for given node 
        if node is self._header or node is self._trailer:
            return None
        else: 
            return self.Position(self,node)
        
    #-----------accessors---------------------------------- 
    def first(self):
        #return first position in the list
        return self._make_position(self._header._next)
    def last(self):
        return self._make_position(self._trailer._prev)
    def before(self,p):
    #return position just before position p
        node = self._validate(p)
        return self._make_position(node._prev)
    def after(self, p):
        node = self._validate(p)
        return self._make_position(node._next)
    def __iter__(self):
        cursor = self.first()
        while cursor is not None:
            yield cursor.element()
            cursor = self.after(cursor)
    
    #----------- mutators ----------------------------------
    def _insert_between(self, e, predecessor, successor):
        #overide inhertied to return position instead of node
        node = super()._insert_between(e, predecessor, successor)
        return self._make_position(node)
    def add_first(self, e):
        return self._insert_between(e, self._header, self._header._next)
    
    def add_last(self,e):
        return self._insert_between(e, self._trailer._prev, self._trailer)
    
    def add_before(self,p,e):
        #insert elm e into list before position p, and return new position
        original = self._validate(p)
        return self._insert_betwen(e, original._prev, original)
    def add_after(self, p, e):
        original = self._validate(p)
        return self._insert_between(e, original, original._next)
    def delete(self, p):
        #remove and return elm at position p
        original = self._validate(p)
        return self._delete_node(original)
    
    def replace(self, p, e):
        # replace elm at position p with e. return former element
        original = self._validate(p)
        old_value = original._element
        original._element = e
        return old_value
```

### Maintaining Access Frequencies

supports:
- access(e): access elm e, increments its access count, and adds to favorites list if not already present
- remove(e): removes e from favorite list
- top(k): returns iteration of the k most accessed elements

composition pattern: single object composed of two or more other objects
- here we use _Item that stores elm and its access count as a single instance and PositionalList, which maintains favorites list

### Advantages to arrays
O(1) time - acess to index. O(k) time to traverse list from beginning
More efficient:
- in linkedQueue, adding elm requires installing new node, changing linkes, and incrementing size. more CPU operations, even though it runs in O(1)
- Array based representations use proprotionall less memory than linked structures. Single linked lists requires 2n refrerences (elm reference and reference to next). Double linked lists used 3n references. dynamiv array in worst case (recently resized, used 2n object references)

### Advantages of link based sequence:
- Linked based strcutres provide worst-case bounds, different than amortized bounds associated with dynamic array
- Linked base strucutres support O(1) time insertions and deletions
    - In python's array list class, inset or pop uses O(n-k+1) time bc of loop to shift all subsequent elements
    



```python
class FavoritesList:
    

  #------------------------------ nested _Item class ------------------------------
  class _Item:
    __slots__ = '_value', '_count'             # streamline memory usage
    def __init__(self, e):
      self._value = e                          # the user's element
      self._count = 0                          # access count initially zero

  #------------------------------- nonpublic utilities -------------------------------
  def _find_position(self, e):
    """Search for element e and return its Position (or None if not found)."""
    walk = self._data.first()
    while walk is not None and walk.element()._value != e:
      walk = self._data.after(walk)
    return walk

  def _move_up(self, p):
    """Move item at Position p earlier in the list based on access count."""
    if p != self._data.first():                      # consider moving...
      cnt = p.element()._count
      walk = self._data.before(p)
      if cnt > walk.element()._count:                # must shift forward
        while (walk != self._data.first() and
               cnt > self._data.before(walk).element()._count):
          walk = self._data.before(walk)
        self._data.add_before(walk, self._data.delete(p))   # delete/reinsert
  
  #------------------------------- public methods -------------------------------
  def __init__(self):
    """Create an empty list of favorites."""
    self._data = PositionalList()                 # will be list of _Item instances

  def __len__(self):
    """Return number of entries on favorites list."""
    return len(self._data)

  def is_empty(self):
    """Return True if list is empty."""
    return len(self._data) == 0

  def access(self, e):
    """Access element e, thereby increasing its access count."""
    p = self._find_position(e)                    # try to locate existing element
    if p is None:
      p = self._data.add_last(self._Item(e))      # if new, place at end
    p.element()._count += 1                       # always increment count
    self._move_up(p)                              # consider moving forward

  def remove(self, e):
    """Remove element e from the list of favorites."""
    p = self._find_position(e)                    # try to locate existing element
    if p is not None:
      self._data.delete(p)                        # delete, if found 

  def top(self, k):
    """Generate sequence of top k elements in terms of access count."""
    if not 1 <= k <= len(self):
      raise ValueError('Illegal value for k')
    walk = self._data.first()
    for j in range(k):
      item = walk.element()                       # element of list is _Item
      yield item._value                           # report user's element
      walk = self._data.after(walk)

  def __repr__(self):
    """Create string representation of the favorites list."""
    return ', '.join('({0}:{1})'.format(i._value, i._count) for i in self._data)

if __name__ == '__main__':
  fav = FavoritesList()
  for c in 'hello. this is a test of mtf':        # well, not the mtf part...
    fav.access(c)
    k = min(5, len(fav))
    print('Top {0}) {1:25} {2}'.format(k, [x for x in fav.top(k)], fav))
        
```
