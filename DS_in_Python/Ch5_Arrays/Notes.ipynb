### Dynamic Arrays
For low-level arrays: exact memory needs to be declared
For tuples and strings, there is no memory constraint since these classes are immutable, so size of array can be fixed when object is instantiated

For Python lists: Python uses abstraction dynamic arrays so that a list has a particular length when constructed, but can add new elements without a limit.

Python list maintains an underlying array with a greater capacity than current lenght of a list. (if user creates it of size 5, it probably is capable of storing 8 references) 

Once additionaly space is used, list class request a new, larger array and initlizes the array so its prefix matches that of the existing array. The old array is reclaimed by the system. 

For list, getsizeof() returns number of bytes devoted to array, not the space devoted to the elements refered to by list.

### Amortized Analysis of Dynamic Arrays

Let S be a sequence of initlal capacity 1 Doubling the array size when full
Total time to preform series of n append operations on S, starting from S being empty is O(n) since each append is O(1)

    Tradeoff is between run-time efficeinty and memory usage. If you double size only to need one extra slot, you waste memory. On the ohter hand, constantly needing to resize wastes time. Key to preformance is keeping amount of additional space proprortional to current size of array

    Ex: adding only 1 additional slot during rezising leads to 1 + 2 + 3... = O(n^2) time. Thats why fixed increments don't work

List uses O(n) memory since when appending its proprotional to overall number of elements. When removing elements, take care to make sure that the dnamic array is still O(n) memory usage


Python Lists doesn't use pure geometric or arithemtic progression, but the append() method uses constant-time behavior
