```python
def prefix_average1(S):
    """Return list such that, for all j, A[j] equals average of S[0], ..., S[j]."""
    n = len(S) #O(1)
    A=[0]* n #create list of zeros O(n)

    for j in range(n): #j = 0 .... n-1 (n times)
        total = 0 #begin comuting S[0] + ... + S[j] O(1)
        for i in range(j + 1):  #exeucted 1 + 2 + .... n times = n(n+1)/2 = O(n^2)
            total += S[i]
        A[j] = total / (j+1) #records average O(1)
    return A

# Calculate run time: O(n) + O(n) + O(n^2) so run time is O(n^2)
```

```python
def prefix_average2(S):
    """Return list such that, for all j, A[j] equals average of S[0], ..., S[j]."""
    n = len(S) #O(1)
    A=[0]* n #create list of zeros O(n)

    for j in range(n): #j = 0 .... n-1 (n times)
        A[j] = sum(S[0:j+1]) / (j+1) #record average still takes 1 + 2 + ... + n time
    return A

#Computation of slice S[0:j+1] takes O(j+1) time as it constructs a new list instance
# Alg is in O(n^2)
```

```python
def prefix_average3(S):
    """Return list such that, for all j, A[j] equals average of S[0], ..., S[j]."""
    n = len(S) #O(1)
    A=[0]* n #create list of zeros O(n)
    total = 0
    for j in range(n): #j = 0 .... n-1 (n times)
        total += S[j]
        A[j] = total / (j+1)
    return A

#For previous examples, prefix sum is computed anew for each val of j
#Here previx sum is maintainged dynamically

#O(n)
```

### Three Way Set Disjointness

Sets A, B, C. No set has duplicate values, but there may be duplicates between sets

Determine the intersection of A B and C 
disjoint if intersection is 0

```python
def disjoint1(A, B, C):
    """ Return True if there is no element common to all three lists.""" 
    for a in A:
        for b in B: 
            for c in C:
                if a == b == c: 
                    return False # we found a common value
    return True  # if we reach this, sets are disjoint
   
   # worse case is O(n^3)
```

```python
def disjoint2(A, B, C):
    """ Return True if there is no element common to all three lists.""" 
    for a in A:
        for b in B: 
            if a == b: #only check C if we found match from A and B
                for c in C:
                    if a == c: 
                        return False # we found a common value
    return True  # if we reach this, sets are disjoint
   
   # wrose case is O(n^2)
   # best case is if A and B are disjoint so its O(n)
```

### Uniqueness Problem

Are there duplicates in Sequence

```python
def unique1(S):
    """ Return True if there are no duplicate elements in sequence S"""
    for j in range(len(S)): #0...n-1
        for k in range(j+1, len(S)):  #1... n-1
            if S[j] == S[k]: #O(1)
                return False # found duplicate pair
    return True # if we reach this, elements were unique

"""
Worst case: (n-1) + n-2 + .... + 1 = O(n^2)
"""
```

```python
# sort S, then look for consecutive duplicates
def unique2(S):
    """Return True if there are no duplicate elements in sequence S."""

    temp = sorted(S) # create a sorted copy of S worse case = O(nlogn)
    for j in range(1, len(temp)): # 1.... n-1 = n-2
        if S[j−1] == S[j]: 
            return False
    return True
# O(n) + O(nlogn) = O(nlogn) time
```

#### Induction and Loop Invariants

Induction: prove statment true for all n by proving n+1

Loop invariants:
    Prove some statement L about a loop is correct
        1. Intital claim Lo is correct before loop begins
        2. If L(j-1) is true before iteration, then Lj is true after iteration j
        3. Final statment Lk implies statement L is true

```python
# use loop-invariant argument to justify that function find() finds the smalles index
# at which val occurs in sequence S
def find(S, val):
    """Return index j such that S[j] == val, or -1 if no such element."""
    n = len(S)
    j=0
    while j < n:
        if S[j] == val: 
            return j
        j += 1 
    return -1
```
