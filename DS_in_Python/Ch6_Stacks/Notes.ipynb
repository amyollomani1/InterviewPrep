FIFO

Array-based implementation is inefficient
- using pop(0) would result in removing the first elm, then shifting everything back, always resulting in worst case O(n)
- To avoid this, don't use pop(0). instead replace dequeed entry with reference to None, and maintain explicit variable f to store the index of elm thats currently at front of queue
    -Another problem: then the size of list wouldn't be proproritonal to number of elements. THIS IS TERRIBLE

Instead use Array Circularly
    Python internal queue class uses three instance variables
        _data: reference to a list instance
        _size: integer representing current num of elements stored
        _front: integer representing first elm of the queue


    ArrayQueue space is not proprotional to O(n) since dequeue never shrinks array. So capacity of array is poroprotional to max number of elements theyre has ever been in the queue, not the current number

    if 0 < self. size < len(self. data) // 4: self. resize(len(self. data) // 2) (USE this two lines at end of dequeue() method so Queue space is O(n))

### Analysis

    Q.enqueue(e): O(1)*
    Q.dequeue(): O(1)*
    Q.first(): O(1)
    Q.is_empty(): O(1)
    len(Q): O(1)

    Space usage is O(n)




```python
class ArrayQueue:
#”””FIFO queue implementation using a Python list as underlying storage.””” 
    DEFAULT_CAPACITY = 10 # moderate capacity for all new queues
    def __init__(self):
    #”””Create an empty queue.”””
        self._data = [None]  *ArrayQueue.DEFAULT_CAPACITY 
        self._size = 0
        self._front = 0
        
    def len(self):
    #”””Return the number of elements in the queue.””” 
        return self. size
    def is_empty(self):
    #”””Return True if the queue is empty.””” 
        return self. size == 0
    def first(self):
    #”””Return (but do not remove) the element at the front of the queue.
        #Raise Empty exception if the queue is empty. ”””
        if self.is_empty():
            raise Empty( 'Queue is empty' ) 
        return self._data[self._front]
    def dequeue(self):
    #”””Remove and return the first element of the queue (i.e., FIFO).
    # Raise Empty exception if the queue is empty. ”””
        if self.is_empty():
            raise Empty( 'Queue is empty' )
        answer = self._data[self._front] #self._front is the indec of the value thats gonna be removed
        self._data[self._front] = None    # help garbage collection removes reference to object
        self._front = (self._front + 1) % len(self._data)  #new first use % for possibility of wrap around
        self._size -= 1
        return answer
 
    def enqueue(self, e):
        #”””Add an element to the back of queue.””” 
        if self._size == len(self._data):
            self._resize(2 * len(self.data)) # double the array size 
        avail = (self. front + self. size) % len(self. data) #this is formular to compute location of back of queue
        self._data[avail] = e
        self._size += 1
    def  _resize(self, cap): # we assume cap >= len(self) 
        #”””Resize to a new list of capacity >= len(self).”””
        old = self._data # keep track of existing list
        self._data = [None] * cap  # allocate list with new capacity
        walk = self._front
        for k in range(self._size): # only consider existing elements
            self._data[k] = old[walk] # intentionally shift indices so front of queue is at index 0
            walk = (1 + walk) % len(old) # use old size as modulus
        self._front = 0 # front has been realigned
```

### Double-Ended Queues

    D.add_first(e): Add elm e to front of deque D
    D.add_last(e): Add elm e to back of D
    D.delete_first(): Remove and return first elm from D
    D.delete_last(): Remove and return last elm of D
    D.first(): Return but don't remove first elm
    D.last(): Return but don't remove last elm
    D.is.empty()
    len(D)

D.add_last(5) [5]
D.add_first(3) [3,5]
D.add_first(7) [7,3,5]
D.first() = 7
D.delete_last() [7,3]
D.delete_first() [3]

                    collections.deque
    D.add_first(e):     D.appendleft()
    D.add_last(e):      D.append()
    D.delete_first():   D.popleft()
    D.delete_last():    D.pop()
    D.first():          D[0]
    D.last():           D[-1]
                        D[j]          Access index
                        D[j] = val    modify arbitary index
                        D.clear()     clear all contents
                        D.rotate(k)   circularly shift rightward k steps
                        D.remove(e)   remove first matching element
                        D.count(e)    count number of matches for e

    Note: calling appendleft() when D is full will result an element from being popped (removed) on the other side

### Analysis
O(1) operations on either end, but O(n) worst case when using index notation near middle of deque
