Recurssion: function makes 1+ calls to itself during execution or by which a data structure relies on smaller instances of very same type of structure in its representation

Examples:

Activation record or frame
* in python, everytime a function is called, a frame is created to store info about the progress of the function
* This record includes a namespace for stroing function call's parameters and local varaibles, and info about which  command in function body is currently executing

* when a function leads to a nested function call, the execution of former func is suspended, and its activation record stores the place in source code taht flow of control should continue upon return of nested call

* Key point: different activation record for each active call


```python
def factorial(n):
    if n == 0:
        return 1 
    else:
        return n * factorial(n-1) 
    
"""
Overall number of operations is O(n) 
since theres n+1 activations and each one is O(1)



"""   
```

```python
# english ruler

def draw_line(tick_length, tick_label= ''    ):
    """ Draw one line with given tick length (followed by optional label)."""
    line =  '-' * tick_length
    if tick_label:
        line += tick_label 
    print(line)
def draw_interval(center_length):
    """ Draw tick interval based upon a central tick length."""
    if center_length > 0:
        draw_interval(center_length- 1) # stop when length drops to 0 
        draw_line(center_length)
        draw_interval(center_length -1)
def draw_ruler(num_inches, major_length):
# recursively draw top ticks
# draw center tick
# recursively draw bottom ticks
#Draw English ruler with given number of inches, major tick length.”””
    draw_line(major_length, 0 )
    for j in range(1, 1 + num_inches): 
        draw_interval(major_length - 1)
# draw inch 0 line
# draw interior ticks for inch # draw inch j line and label
        draw_line(major_length, str(j))
        
draw_ruler(3,10)

"""
Number of lines is  2^c -1


"""
```

### Binary Search

Use when sequence is sorted, otherwise use sequential search on unsorted list

Explanation
* mid = floor((low+high)/2)
* if target = data(mid) then search terminates
* if target < data[mid], then recur first half pf seauence from low to mid-1
* if target > data [mid] then recur on second half : indicies from mid + 1 to high

Unsuccesful search if low > high, since interval [low,high] is empty


```python
def binary_search(data, target, low, high):
#”””Return True if target is found in indicated portion of a Python list.
#The search only considers the portion from data[low] to data[high] inclusive. ”””
    if low > high:
        return False 
    else:
        mid = (low + high) // 2 
        if target == data[mid]:
            return True
        elif target < data[mid]:
    # recur on the portion left of the middle
    # interval is empty; no match # found a match
            return binary_search(data, target, low, mid - 1) 
        else:
            # recur on the portion right of the middle
            return binary_search(data, target, mid + 1, high)
        
        
""" 
O(logn) for sorted list
mac number of recursive calls preformed is smallest int r such that
n/(2^r) is < 1

r = floor(logn) + 1

"""
```

### Anaylizing recursive algs

Recusrion trace: amount of recursive activations occur

amoritzation: a tighter bound on a series of opeartions instead of assuming each achieves the worse case

Infinite recursion: if each recursive call makes another recursive call without reaching a base case is a fatal error.

In Python, you can only have 1000 functions simultaneously active.
    If this limit is reached, Python interpreter raises RuntimeError with a message: 'maximum recursion depth exceeded'

Above binary alg has O(logn) recursive depth, so there would need to be 2^1000 elm for limit to be reached

To change limit call
sys.setrecursionlimit(1000000)

reflects structure of recursion trace. not running time:
Linear recursion: recursive call starts at most one other call (ex:binary search)
Binary recursion; recursive call may start two others
Multiple recursion: recursive call may start 3+ others

#### Designing recursive algs:
1. test for base cases. Handling base cases shouldn't use recurssion
2. Recur. Define 1+ recursive calls so it makes progress towards base case

Pros and Cons to recursion:
Pro: simple design thats repetive can avoid complex if statments and loops, making alg more readable
Con: Python interpreter needs to maintain activation recrods to keep track of state of each nested call, bad for memory
    Fix: use stack in nonrecursive alg to maintain nested recursive structure ourselves. This shifts memory usage from interpreter to out stack, which reduces memory usage by storing only minimal info necessary

Tail recurssion: eliminate recursion without using axillary memory
    - recursion is a tail recurssion if any recursive call made from one context is very las operation in that contex, so the return value of the recursive call (if any) immedialy returned by the envlosing recursion.
    - Must be a linear recursion
    - binary_search and reverse() are examples of tail recursions
    - factiorail() isn't tail recursion since it concludes with return n* factorial(n-1) command. Not tail since additional multiplication is preformed after recursive call is completed



```python
def binary_search_iterative(data, target):
#”””Return True if target is found in the given Python list.””” 
    low = 0
    high = len(data)-1
    while low <= high:
        mid = (low + high) // 2 
        if target == data[mid]:
            return True
        elif target < data[mid]:
            high = mid - 1 
        else:
            low = mid + 1
    return False

#non recursive implementation
```

```python
def linear_sum(S, n):
    #return sum of first n numbers of S
    if n == 0:
        return 0
    else:
        return linear_sum(S, n-1) + S[n-1]
    
#Memory space is O(n)
#Theres n+1 activation records
#so it Takes O(n) time
```

```python
def reverse(S,start, stop):
    #Reverse elements in implicit slice
    if start < stop -1:
        S[start], S[stop - 1] = S[stop - 1], S[start]
        reverse(S, start + 1, stop - 1)
   
   # total of 1 + floor(n/2) recursive calls
   #Each call is O(1) so process is O(n)
    
```

```python
def reverse_iterative(S):
#”””Reverse elements in sequence S.””” 
    start, stop = 0, len(S)
    while start < stop - 1:
        S[start], S[stop-1] = S[stop-1], S[start] 
        start, stop = start + 1, stop - 1
        
#start stop values updated at each pass
#
```

```python
def power(x,n):
    if n == 0:
        return 1
    else:
        return x* power(x,n-1)
    
#O(n) time
```

```python
def better_power(x,n):
    if n == 0:
        return 1 
    else:
        partial = power(x, n // 2) 
        result = partial * partial
    if n % 2 == 1:
        result  *= x 
        return result
    
#imrpvoe: O(logn) and memory is O(logn)
```

```python
# Element Uniqueness problem

def unique3(S, start, stop):
    if stop - start <= 1: return True
    elif not unique3(S, start, stop -1): return False
    elif not unique3(S, start+1, stop): return False
    else: return S[start] != S[stop-1]
    
"""
    nonreucrisve part of each call uses O(1) time
so if n = 1, operates in O(1)
each call can result in two recursive calls.
so n results in two calls to n-1, which goes to four calls of n-2 and so forth
1 + 2 + 4 +.... 2^(n-1) = 2^n - 1
O(2^n)

"""
```

```python
# Bad Fibonocci
def bad_fibonacci(n):
    if n<= 1:
        return n
    else:
        return bad_fibonacci(n-2) + bad_fibonacci(n-1)
    
#exponential time
```

```python
def good_fibonacci(n):
    if n<=1 :
        return(n,0)
    else:
        (a,b) = good_fibonacci(n-1)
        return(a+b,a)

#recursion trace include n function calls, each at constant time
#O(n) overall time



    
```
